# Histograms or boxplots to visualize the distributions of various numerical variables.

```python
cat_cols=df.select_dtypes(include=['object']).columns
num_cols = df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(df[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    df[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.show()
```

The dataset contains variables with a mix of normal, skewed, and uniform distributions.
Many columns exhibit skewness (either left or right), indicating that transformations might be needed for certain analyses. 
Outliers are present in several columns, which could impact the results of statistical analyses and may need to be addressed.

# Bar chart

```python
fig, axes = plt.subplots(7, 2, figsize=(50, 50))
fig.suptitle('bar plot for all categorical variables in the dataset')
sns.countplot(ax=axes[0, 0], x='LB', data=df, color='blue', order=df['LB'].head(20).value_counts().index)
sns.countplot(ax=axes[0, 1], x='AC', data=df, color='blue', order=df['AC'].head(8).value_counts().index)
sns.countplot(ax=axes[1, 0], x='FM', data=df, color='blue', order=df['FM'].head(10).value_counts().index)
sns.countplot(ax=axes[1, 1], x='UC', data=df, color='blue', order=df['UC'].head(10).value_counts().index)
sns.countplot(ax=axes[2, 0], x='DL', data=df, color='blue', order=df['DL'].head(8).value_counts().index)
sns.countplot(ax=axes[2, 1], x='DS', data=df, color='blue', order=df['DS'].head(80).value_counts().index)
sns.countplot(ax=axes[3, 0], x='DP', data=df, color='blue', order=df['DP'].head(80).value_counts().index)
sns.countplot(ax=axes[3, 1], x='ASTV', data=df, color='blue', order=df['ASTV'].head(80).value_counts().index)
sns.countplot(ax=axes[4, 0], x='MSTV', data=df, color='blue', order=df['MSTV'].head(80).value_counts().index)
sns.countplot(ax=axes[4, 1], x='ALTV', data=df, color='blue', order=df['ALTV'].head(80).value_counts().index)
sns.countplot(ax=axes[5, 0], x='MLTV', data=df, color='blue', order=df['MLTV'].head(80).value_counts().index)
sns.countplot(ax=axes[5, 1], x='Width', data=df, color='blue', order=df['Width'].head(80).value_counts().index)
sns.countplot(ax=axes[6, 0], x='Tendency', data=df, color='blue', order=df['Tendency'].head(80).value_counts().index)
sns.countplot(ax=axes[6, 1], x='NSP', data=df, color='blue', order=df['NSP'].head(80).value_counts().index)
# Adjusting tick parameters for better readability
axes[1, 1].tick_params(labelrotation=45)
axes[2, 0].tick_params(labelrotation=90)
axes[2, 1].tick_params(labelrotation=90)

plt.tight_layout()
plt.show()
```

By observing the peaks and distributions in each count plot, you can identify which categories are dominant and which are rare. 
This helps in understanding the balance of your dataset and in making decisions for further analysis or model training.

# Heat Map for the correlation between the variables, whether it has a positive or negative correlation.

```python
plt.figure(figsize=(12, 7))
sns.heatmap(df.drop(['LB','LB'],axis=1).corr(), annot = True, vmin = -1, vmax = 1)
plt.show()
```

Variables width and MSTV have a correlation coefficient of 0.44, indicating they are strongly positively correlated.
Variables ALTV and Width have a correlation coefficient of -0.35, indicating a strong negative correlation.
Variables width and UC have a correlation coefficient close to 0.1, indicating little to no linear relationship.
